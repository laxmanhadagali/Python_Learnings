**LOOPS IN PYTHON**

```python
#there are two types of loops
# for loop
#while loop
'''3 main components of while
1.initialization
2.condition
3.increment/update'''

#simple while loop
i=1
while i<=10:
    print("Hello")
    i=i+1
    
```

```python
i=1
while i<=20:
    print(i)
    i=i+1
```

```python
#break and continue
i=1
while i<=10:
    if i>5:
       break 
    print(i)
    i=i+1
```

```python
i=1
while i<=10:
    if i==10:
        break
    print(i)
    i=i+1
```

```python
i=1
while i<=10:
    i=i+1
    if i==5:
        continue
    print(i)
```

```python
#quiz
quest1="who is the prime minister of india"
c_answer="Narendra Modi"
print(quest1)
user_ans=input("enter the answer")
attempts=0
    
while c_answer!=user_ans:
    attempts=attempts+1
    if attempts==3:
        print("You are out of attempts")
        break
    print("you are wrong try again")
    user_ans=input()
else:
    print("You are right")
    
```

```python
names=['laxman','chetan','vishwa','riyaz']
upper_case=[]
i=0
while i<len(names):
    upper_case.append(names[i].upper())
    i=i+1
print(upper_case)    
    
```

**FOR LOOP**

```python
names=['laxman','chetan','vishwa','riyaz']
for i in names:
    print(i)
```

```python
for i in range(1,20,2):
    print(i)
```

```python
#simple expense program
expenses=[200,100,25,135,300,40]
sum=0
for x in expenses:
    sum=sum+x
print(sum)
del sum
```
