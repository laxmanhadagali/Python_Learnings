```python
#Filter function
#finding odd_numbers
numbers=[1,2,3,4,5,6,7,8]
odd_numbers=list(filter(lambda c:c%2!=0,numbers))
print(odd_numbers)
```

```python
#finding even numbers
numbers=[1,2,3,4,5,6,7,8]
even_numbers=list(filter(lambda c:c%2==0,numbers))
print(even_numbers)
```

```python
#MAP FUNCTION : map all the numbers and gives the output
numbers=[1,2,3,4,5,6,7,8]
square=list(map(lambda x:x*x,numbers))
print(square)
```

```python
#REDUCE FUNCTION : help to reduce the 
from functools import reduce
numbers=[1,2,3,4,5,6]
total=reduce(lambda a,b:a+b,numbers)
print(total)
```

```python
#DICE Game 
import random
roll=lambda:random.randint(1,6)

user1=input("enter the user1 name")
user2=input("enter the user2 name")

v1=roll()
v2=roll()
print(user1,"rolled",v1)
print(user2,"rolled",v2)

if v1>v2:
    print(user1,"wins")
elif v2>v1:
    print(user2,"wins")
else:
    print("It's a tie")


```
