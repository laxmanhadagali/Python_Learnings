**PYTHON MODULES and PACKAGES**

```python
#modules and packages are imported using import keyword
#example
import math
number=math.sqrt(225)
print(number)

#from ....import statement
#importing the sqrt function directly from the math module
from math import sqrt
#using the sqrt functiondirectly
result=sqrt(25)
print(result)

#multiple objects or functions can be imported using ,,
from math import sqrt,sin,cos
result1=sqrt(25)
result2=sin(0.5)
result3=cos(1.2)
print(result1)
print(result2)
print(result3)
```

```python
#the import as statement (allow us to import a module assign it an alias providing a shorter and more convinient way refer to module)
import math as m
result=m.sqrt(625)
print(result)

#SCHEMA EXAMPLE(use random module and generatw a random number with a specified range)
import random
#generate a random number between 1 and 10
random_number=random.randint(1,10)
print(random_number)


```

```python
#geenerating a random numbers,selecting random elements and shuffling sequence

import random
#select a random element from a list
my_list=['apple','banana','cherry','durain']
random_element=random.choice(my_list)
print(random_element)
#shuffle a list
random.shuffle(my_list)
print(my_list)
```
