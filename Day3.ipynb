**Date and Time PACKAGES**

#PACKAGES ARE FOLDER WHICH CONSISTS OF MULTIPLE MODULE
#date and time packages are help us to work with date and time
#it has 4 properties
- .date #year month day
- 2.time #Hour Minute second milisecond
- 3.datetime #Date+time togrther
- .timedelta #gap between two dates or time

```python
from datetime import date
my_date=date(2001,6,13) #simple date creation
print(my_date)
Today_date=date.today()
print(Today_date.year)
print(Today_date.month)
print(Today_date.day)
print(Today_date)
print(Today_date.isoweekday())
```

```python
from datetime import datetime as dt
date1=dt.now()
print(date1)
```

```python
from datetime import time
my_time=time(7,8,45)#simple time creation
print(my_time)
```

```python
from datetime import timedelta
cd=date.today()
fd_after2y=cd+timedelta(days=730)
print(fd_after2y)
```
