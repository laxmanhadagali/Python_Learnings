### Variables in python
* A variable in Python is a name that stores a value.
* It acts like a container that holds data which can be used and modified throughout a program.

#### Creating And Printing Varaibles

```python
# Creating variables
name = "Laxman"
age = 21
city = "Hubli"

# Printing variables
print(name)
print(age)
print(city)
```

```python
# Checking the data type using type()
print(type(name))   # string
print(type(age))    # integer
print(type(city))   # string
```

```python
# Assigning multiple values at once
a, b, c = 10, 20, 30
print(a, b, c)

# Assigning one value to multiple variables
x = y = z = 50
print(x, y, z)
```

```python
# Valid variable names
student_name = "John"
studentAge = 21
_student = "valid"

# Invalid variable names (Try running one by one)
# 1name = "Error"   # starts with number 
# student-name = "Error"  # contains hyphen 
# class = "Error"   # reserved keyword
```

```python
# Reassigning values to the same variable
fruit = "Apple"
print(fruit)

fruit = "Mango"   # changed value
print(fruit)
```

```python
name = "Laxman"
age = 21

# Using f-string
print(f"My name is {name} and I am {age} years old.")

# Using format() method
print("My name is {} and I am {} years old.".format(name, age))
```

```python
a = 10
b = 5

# Performing arithmetic operations
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Power:", a ** b)
```

```python
x = 10
y = 20

# Swapping using a temporary variable
temp = x
x = y
y = temp
print("After swapping:", x, y)

# Swapping without a temporary variable
a, b = 5, 15
a, b = b, a
print("After swapping:", a, b)
```

```python
PI = 3.14159
GRAVITY = 9.8
print("Pi value:", PI)
print("Gravity:", GRAVITY)
```

```python
# Taking user input
name = input("Enter your name: ")
age = int(input("Enter your age: "))

print(f"Hello {name}, you are {age} years old!")
```

### Datatypes in python
* Datatypes in python are the type of data that a variable can hold

**1.Integer**

```python
a=10
b=15
print(a+b)
```

```python
#floating point
x=3.52
y=1.48
print(x+y)
```

```python
#complex numbers
c=2+3j
d=1-2j
print(c*d)
```

**2.Strings**

```python
#concatenation
First_Name="Laxman"
Last_Name="Hadagali"
Full_Name=First_Name+" "+Last_Name
print(Full_Name)
```

```python
#string slicing
message="Hello World"
print(message[0:4])
print(message[0:5])
print(message[0:7])
print(message[0:12])
```

```python
#string formatting
age=21
print(f"I am {age} years old")
```

**3.Lists**


```python
#creating a list
numbers=[1,2,3,4,5]
#Acessing elements
print(numbers[0])
print(numbers[4])
print(numbers[3])
```

```python
#modyfying elements
numbers[0]=10
print(numbers[0])
```

```python
#adding elemnts
numbers.append(6)
print(numbers)
#Removing elements
numbers.remove(3)
print(numbers)
```

**4.Tuples**

```python
#creating a tuple
coordinates=(0,1)
#accessing a tuple
x=coordinates[0]
y=coordinates[1]
print("x=",x)
print("y=",y)
```

**5.Sets**

```python
#creating a set 
fruits={"Orange","banana","apple"}
#adding elements
fruits.add("grape")
print(fruits)
fruits.remove("banana")
print(fruits)
```

**6.Dictionaries**

```python
#creating a dictionary
person={

    "name":"John Doe",
    "age":25,
    "city":"London"
}

#accesing Value
print(person["name"])
print(person.get("age"))

#Modyfying values
person["age"]=30
print(person)
```
