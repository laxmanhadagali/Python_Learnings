**TUPLES**

```python
#Tuple is craeted using ()
tup1=()
type(tup1)
```

```python
tup2=(1,2,3,4) #tuple of integers
tup3=("Banana","orange","apple") #tuple of strings
tup4=(10.5,2,5,4.7)#tuple of float
tup5=(1,2.5,"Lucky") #tuple of mixed datatype
tup6=(1,5,(2,6),(3.5,1.8),("Mrshannu","Teju","Raghu")) #tuple inside tuple
print(tup1)
print(tup2)
print(tup3)
print(tup4)
print(tup5)
print(tup6)
```

```python
#indexing
print(tup2[1])
print(tup3[2])
print(tup6[4][1])
print(tup5[2])
print(tup6[4][1:])
```

```python
tup7=("Monday","Tuesday","Wednesday","Thursday","Friday","Sturday","Sunday")
print(tup7[5:])
```

```python
#immutable we can delete entire tuple but we cant change the tuple
#tup2[0]=30
del tup2
```

```python
tup8=("One","Two","One","Three")
tup8.count("One")
```

```python
#indexing
tup8.index('Two')
```

```python
#membership
print("Three" in tup8)
print("five" in tup8)
```

```python
#sorting we have pass the the tuple as parameter
tup9=(9,8,1,7,4,6,10,2,3)
print(sorted(tup9))
print(sorted(tup9,reverse=True))
```

**SETS**

```python
#sets are created usng curly/flower brackets
myset=set() #empty set
type(myset)
```

```python
set1={1,2,1,2,1,3,4,2,3,5}
print(set1)
```

```python
set2={8,9,5,8,6,5,4,5,3,}
print(set2)
```

```python
set3={"TCS","TechMahindra","Wipro","Cognizant"} #set of strings
set4={3,4.0,'Magic'} #set od mixed datatype
```

```python
#membership
print("Wipro" in set3)
print("Delloite" in set3)
```

```python
#adding and updating the  valule
set4.add(10)
set4.update(['IronMan','Spiderman',17])
print(set4)
```

```python
#remove and discard 
set4.remove(17)
set4.discard(3)
```

```python
print(set4.clear())
```
